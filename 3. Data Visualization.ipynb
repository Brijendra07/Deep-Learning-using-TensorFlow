# Matplotlib basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0,1,2]
y = [100,200,300]
```

```python
plt.plot(x,y)
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
'price':[100,120,190,200,230,310,330,305]})
```

```python
housing
```

```python
plt.scatter(housing['rooms'],housing['price'])
```

```python
plt.plot(x,y,color = 'red',marker='o')
plt.xlim(0,2)
plt.ylim(0,300)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

# Seaborn basics

```python
df = pd.read_csv('DATA/heart.csv')
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,8))
sns.distplot(df['age'],color='red')
```

```python
sns.countplot(x='sex',data=df,hue='cp')
```

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.scatterplot(x='chol',y = 'trestbps',data=df,hue = 'sex')
```

```python
iris = pd.read_csv('DATA/iris.csv')
```

```python
sns.pairplot(iris,hue = 'species')
```
